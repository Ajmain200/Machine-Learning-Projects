```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("/content/AI-ML__Dataset_PoC - payments.csv")
df_items = pd.read_csv("/content/AI-ML__Dataset_PoC - item.csv")

df["name"] = df["item_oid"].map(df_items.set_index("oid")["name"])

cost_cols = ["cd", "rd", "sd", "vat"]
for col in cost_cols:
    df[col] = df[col].str.replace(",", "").astype(float)

df["total_cost"] = df[cost_cols].sum(axis=1)

SELECTED_ITEM = "itm-20000"

if SELECTED_ITEM.lower() != "all":
    df_item = df[df["item_oid"] == SELECTED_ITEM].copy()
    if df_item.empty:
        raise ValueError(f"Selected item '{SELECTED_ITEM}' not found in dataset")
else:
    df_item = df.copy()


yearly_cost = df_item.groupby("YearValue", as_index=False)["total_cost"].sum()

X = yearly_cost[["YearValue"]]
y = yearly_cost["total_cost"]

model = LinearRegression()
model.fit(X, y)

def predict_item_yearly_cost(year):
    X_input = pd.DataFrame([[year]], columns=["YearValue"])
    return model.predict(X_input)[0]

year = 2026
predicted_cost = predict_item_yearly_cost(year)

item_label = "all items" if SELECTED_ITEM.lower() == "all" else SELECTED_ITEM
print(f"Predicted yearly total cost for {item_label} in {year}: {predicted_cost}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("/content/AI-ML__Dataset_PoC - payments.csv")
df_items = pd.read_csv("/content/AI-ML__Dataset_PoC - item.csv")

df["name"] = df["item_oid"].map(df_items.set_index("oid")["name"])

cost_cols = ["cd", "rd", "sd", "vat"]
for col in cost_cols:
    df[col] = df[col].str.replace(",", "").astype(float)

df["total_cost"] = df[cost_cols].sum(axis=1)


selected_item = input("Enter item_oid or 'all': ").strip()
year = int(input("Enter year to predict: "))

if selected_item.lower() != "all":
    df_item = df[df["item_oid"] == selected_item].copy()
    if df_item.empty:
        raise ValueError(f"Item '{selected_item}' not found")
else:
    df_item = df.copy()

yearly_cost = df_item.groupby("YearValue", as_index=False)["total_cost"].sum()

X = yearly_cost[["YearValue"]]
y = yearly_cost["total_cost"]

model = LinearRegression()
model.fit(X, y)

X_input = pd.DataFrame([[year]], columns=["YearValue"])
predicted_cost = model.predict(X_input)[0]

item_label = "all items" if selected_item.lower() == "all" else selected_item
print(f"Predicted total cost for {item_label} in {year}: {predicted_cost}")
```
